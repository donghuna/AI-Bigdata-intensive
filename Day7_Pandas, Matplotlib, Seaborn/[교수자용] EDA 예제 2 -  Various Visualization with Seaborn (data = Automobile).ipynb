# Example 2: Various Visualization with Seaborn

```python
# 패키지 import
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore') #Warning Off
```

```python
# 데이터 로드
auto = pd.read_csv('Automobile.csv')
auto.head()
```

```python
# 분포 그래프 출력해보기 (변수 = normalized_losses)
sns.distplot(auto['normalized_losses'])
plt.show()
```

```python
# jointplot 그래프 출력해보기 (변수 = engine_size, horsepower)
sns.jointplot(data = auto, x = "engine_size", y = "horsepower")
plt.show()
```

```python
# 다른 종류의 그래프 형태 확인해보기
# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }
sns.jointplot(data = auto, x = "engine_size", y = "horsepower", kind = 'hex')
plt.show()
```

```python
# pairplot
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']]);
```

```python
# stripplot
sns.stripplot(data = auto, x = 'fuel_type', y = 'horsepower', jitter = True);
```

```python
# BoxPlot
sns.boxplot(data = auto, x = auto['number_of_doors'], y = auto['horsepower'], hue=auto['fuel_type']);
```

```python
# BarPlot
sns.barplot(data = auto, x = auto['body_style'], y = auto['horsepower'], hue= auto['engine_location']);
```

```python
# count plot
sns.countplot(data = auto, x = auto['body_style'], hue=auto['engine_location']);
```

```python
# 선형적합 그래프
sns.lmplot(x='horsepower', y ='peak_rpm', data= auto, hue='fuel_type');
```
