# Example 1: College Football Data in U.S.

```python
# 패키지 import
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore') # Warning Off
```

```python
# 데이터 로드
df = pd.read_csv('CFB2019.csv')
```

```python
# 데이터 구조 살펴보기
display(df) #Initial structure of the data
```

```python
# 전체 데이터 요약 통계 출력하기
df.describe()
```

```python
# 전체 데이터 정보 파악하기
print("* 원 데이터 크기: ", df.shape)
print("")
print("* 변수 리스트:")
print(df.columns.values)
print("")
print("* 데이터 정보 요약 출력:")
print(df.info)
print("")
print("* 데이터 타입 확인 출력:")
print(df.dtypes)
```

```python
# 팀 리스트 출력
team = df['Team']
print(team)
```

```python
# 상관계수 행렬 구하기
cor = df.corr()
display(cor)
```

```python
# Win 변수가 0.75 이상인 변수만 선택하여, Win 변수와의 상관계수 열만 선택 후, columns_list에 해당 변수 리스트 저장
columns = cor[cor["Win"]>0.75]['Win']
columns_list = columns.index
```

```python
# Win 변수가 0.75 이상인 변수 데이터만 선택하여 df1에 저장
df1 = pd.DataFrame(df, columns=columns_list)
display(df1)
```

```python
# Games열은 제거하고, Team이라는 이름의 새로운 열에 팀 명칭을 입력
df2 = df1.drop('Games', axis= 1)
df2['Team'] = team
display(df2)
```

```python
# Seaborn 패키지를 활용하여 히트맵 생성
# 차트(fig)의 크기 설정
plt.figure(figsize=(14,12))
# 히트맵 생성
sns.heatmap(df2.corr(),linewidths=.1,cmap="YlGnBu", annot=True) # annot=True -> 히트맵에 값을 함께 출력
# y축의 라벨 글자를 회전 (기본상태 = 90, x축과 같은 방향 = 0)
plt.yticks(rotation=0);
```

```python
# Seaborn 패키지를 활용하여 pairplot (scatter plot matrix) 생성
g = sns.pairplot(df2)
```

```python
# Plot colored by continent for years 2000-2007
sns.pairplot(df2[df2['Win'] > 5], 
             vars = ['Off.Yards', 'Touchdowns', 'PAT'], diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, # 이변량 부분 (산점도)의 외관 설정
             size = 4);
# Title 
plt.suptitle('Pair Plot of influential factors for winning', y=1.05, size = 28);
```
