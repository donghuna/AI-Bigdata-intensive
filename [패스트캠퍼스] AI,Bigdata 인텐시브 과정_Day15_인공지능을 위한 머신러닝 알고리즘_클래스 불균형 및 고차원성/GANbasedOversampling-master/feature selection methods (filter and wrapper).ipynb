##Filter-based method (Example)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

def evaluate_feature_selection_methods(X, y, methods, clf, cv):
    results = []
    for method_name, method_obj in methods:
        acc_scores = []
        recall_scores = []
        for k in range(1, method_obj.shape[1]+1):
            X_sel = method_obj[:, :k]
            acc_score = np.mean(cross_val_score(clf, X_sel, y, cv=cv))
            recall_score = np.mean(cross_val_score(clf, X_sel, y, cv=cv, scoring='recall'))
            acc_scores.append(acc_score)
            recall_scores.append(recall_score)
        results.append((method_name, acc_scores, recall_scores))
    return results


# Generate a balanced dataset with 5000 samples and 1000 features
X_bal, y_bal = make_classification(n_samples=5000, n_features=1000, n_informative=10,
                                     n_redundant=0, n_clusters_per_class=2, weights=[0.5, 0.5],
                                     random_state=42)

# Generate an imbalanced dataset with 5000 samples and 1000 features
X_imb, y_imb = make_classification(n_samples=5000, n_features=1000, n_informative=10,
                                     n_redundant=0, n_clusters_per_class=2, weights=[0.9, 0.1],
                                     random_state=42)

# Scale the features to the range [0, 1]
scaler = MinMaxScaler()
X_bal = scaler.fit_transform(X_bal)
X_imb = scaler.fit_transform(X_imb)

# Define the feature selection methods
corr = pd.DataFrame(X_bal).corr(method='pearson').abs()
relevant_features = corr[corr > 0.2]
relevant_features = relevant_features.sum()[(relevant_features.sum() > 1)].index.tolist()
methods = [('Correlation-based', X_bal[:, relevant_features]),
           ('Mutual information-based', SelectKBest(mutual_info_classif, k=10).fit_transform(X_bal, y_bal)),
           ('Chi-squared', SelectKBest(chi2, k=10).fit_transform(X_bal, y_bal))]

# Define the classifier and the number of cross-validation folds
clf = KNeighborsClassifier(n_neighbors=5)
cv = 5

# Evaluate the performance of each feature selection method with different feature subset sizes
results_balanced = evaluate_feature_selection_methods(X_bal, y_bal, methods, clf, cv)
results_imbalanced = evaluate_feature_selection_methods(X_imb, y_imb, methods, clf, cv)
```

```python
# Define the x-axis tick labels
x_tick_labels = list(range(1, methods[0][1].shape[1]+1))

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Define the plot data
plot_data = [results_balanced, results_balanced, results_imbalanced, results_imbalanced]
plot_titles = ['Comparison in accuracy for balanced datasets',
               'Comparison in recall for balanced datasets',
               'Comparison in accuracy for imbalanced datasets',
               'Comparison in recall for imbalanced datasets']
y_labels = ['Accuracy', 'Recall', 'Accuracy', 'Recall']

# Iterate over the subplots and plot the data
for i in range(4):
    ax = axs[i // 2, i % 2]
    ax.set_title(plot_titles[i])
    ax.set_xlabel('Number of features')
    ax.set_ylabel(y_labels[i])
    if i%2 == 0 :
        for method_idx in range(len(methods)):
            method_name, metric_scores = plot_data[i][method_idx][0], plot_data[i][method_idx][1][:7]
            ax.plot(x_tick_labels[:7], metric_scores, label=method_name)
    else:
        for method_idx in range(len(methods)):
            method_name, metric_scores = plot_data[i][method_idx][0], plot_data[i][method_idx][2][:7]
            ax.plot(x_tick_labels[:7], metric_scores, label=method_name)
    ax.legend()

# Show the plot
plt.show()
```

##Wrapper-based method (Example)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LassoCV
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

def evaluate_feature_selection_methods(X, y, methods, clf, cv):
    results = []
    for method_name, method_obj in methods:
        acc_scores = []
        recall_scores = []
        for k in range(1, method_obj.shape[1]+1):
            X_sel = method_obj[:, :k]
            acc_score = np.mean(cross_val_score(clf, X_sel, y, cv=cv))
            recall_score = np.mean(cross_val_score(clf, X_sel, y, cv=cv, scoring='recall'))
            acc_scores.append(acc_score)
            recall_scores.append(recall_score)
        results.append((method_name, acc_scores, recall_scores))
    return results

# Generate a balanced dataset with 5000 samples and features
X_bal, y_bal = make_classification(n_samples=5000, n_features=20, n_informative=10,
                                     n_redundant=0, n_clusters_per_class=2, weights=[0.5, 0.5],
                                     random_state=42)

# Generate an imbalanced dataset with 5000 samples and features
X_imb, y_imb = make_classification(n_samples=5000, n_features=20, n_informative=10,
                                     n_redundant=0, n_clusters_per_class=2, weights=[0.9, 0.1],
                                     random_state=42)

# Scale the features to the range [0, 1]
scaler = MinMaxScaler()
X_bal = scaler.fit_transform(X_bal)
X_imb = scaler.fit_transform(X_imb)

# Define the feature selection methods
forward_selection = SequentialFeatureSelector(clf, n_features_to_select=10, direction='forward').fit(X_bal, y_bal)
selected_features = forward_selection.get_support(indices=True)
methods = [('Forward selection', X_bal[:, selected_features])]
```

```python
# Define the classifier and the number of cross-validation folds
clf = KNeighborsClassifier(n_neighbors=5)
cv = 5

# Evaluate the performance of each feature selection method with different feature subset sizes
results_balanced = evaluate_feature_selection_methods(X_bal, y_bal, methods, clf, cv)
results_imbalanced = evaluate_feature_selection_methods(X_imb, y_imb, methods, clf, cv)
```

```python
# Define the x-axis tick labels
x_tick_labels = list(range(1, methods[0][1].shape[1]+1))

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Define the plot data
plot_data = [results_balanced, results_balanced, results_imbalanced, results_imbalanced]
plot_titles = ['Comparison in accuracy for balanced datasets',
               'Comparison in recall for balanced datasets',
               'Comparison in accuracy for imbalanced datasets',
               'Comparison in recall for imbalanced datasets']
y_labels = ['Accuracy', 'Recall', 'Accuracy', 'Recall']

# Iterate over the subplots and plot the data
for i in range(4):
    ax = axs[i // 2, i % 2]
    ax.set_title(plot_titles[i])
    ax.set_xlabel('Number of features')
    ax.set_ylabel(y_labels[i])
    if i%2 == 0 :
        for method_idx in range(len(methods)):
            method_name, metric_scores = plot_data[i][method_idx][0], plot_data[i][method_idx][1][:7]
            ax.plot(x_tick_labels[:7], metric_scores, label=method_name)
    else:
        for method_idx in range(len(methods)):
            method_name, metric_scores = plot_data[i][method_idx][0], plot_data[i][method_idx][2][:7]
            ax.plot(x_tick_labels[:7], metric_scores, label=method_name)
    ax.legend()

# Show the plot
plt.show()
```
