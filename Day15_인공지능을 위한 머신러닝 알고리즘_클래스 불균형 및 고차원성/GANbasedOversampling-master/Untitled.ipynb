```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Define a function to get the sentiment score of a sentence
def get_sentiment_score(sentence):
    return sia.polarity_scores(sentence)['compound']

# Define some example sentences to analyze
sentences = [
    "I love this movie!",
    "This restaurant is terrible.",
    "The weather is perfect today.",
    "I feel sad about what happened.",
    "She is the best singer in the world!",
    "I hate you"
]

# Analyze each sentence and print the results
for sentence in sentences:
    score = get_sentiment_score(sentence)
    if score > 0:
        print(f"{sentence} -> Positive")
    elif score < 0:
        print(f"{sentence} -> Negative")
    else:
        print(f"{sentence} -> Neutral")
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Initialize the sentiment analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Define a function to get the sentiment score of a sentence
def get_sentiment_score(sentence):
    return sia.polarity_scores(sentence)['compound']

# Define some example sentences to analyze
sentences = [
    "I love this movie!",
    "This restaurant is terrible.",
    "The weather is perfect today.",
    "I feel sad about what happened.",
    "She is the best singer in the world!"
]

# Analyze each sentence and plot the results
scores = []
for sentence in sentences:
    score = get_sentiment_score(sentence)
    scores.append(score)
    if score > 0:
        print(f"{sentence} -> Positive")
    elif score < 0:
        print(f"{sentence} -> Negative")
    else:
        print(f"{sentence} -> Neutral")

plt.bar(range(len(scores)), scores)
plt.xticks(range(len(sentences)), sentences)
plt.xlabel('Sentences')
plt.ylabel('Sentiment Score')
plt.show()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Initialize the sentiment analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Define a function to get the sentiment score of a sentence
def get_sentiment_score(sentence):
    return sia.polarity_scores(sentence)['compound']

# Define some example sentences to analyze
sentences = [
    "I love this movie!",
    "This restaurant is terrible.",
    "The weather is perfect today.",
    "I feel sad about what happened.",
    "She is the best singer in the world!"
]

# Analyze each sentence and plot the results
scores = []
colors = []
for sentence in sentences:
    score = get_sentiment_score(sentence)
    scores.append(score)
    if score > 0:
        colors.append('b')  # blue for positive sentences
        print(f"{sentence} -> Positive")
    elif score < 0:
        colors.append('r')  # red for negative sentences
        print(f"{sentence} -> Negative")
    else:
        colors.append('k')  # black for neutral sentences
        print(f"{sentence} -> Neutral")

plt.bar(range(len(scores)), scores, color=colors)
plt.xticks(range(len(sentences)), sentences)
plt.xlabel('Sentences')
plt.ylabel('Sentiment Score')
plt.show()

```
