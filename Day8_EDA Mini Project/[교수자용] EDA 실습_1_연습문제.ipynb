### Problem 1
주어진 리스트에 대해 five number summary 값을 계산하고 출력하시오.

```python
score = [1,3,5,5,7,9,11,11,12,14,15,16,16,16,16,17,17,17,18,20,21,22,23,24,25]
f_summary = np.percentile(score, [0, 25, 50, 75, 100])

print("="*25)
print(" Five Number Summary")
print(" (data: score_list)")
print("="*25)
print(" Min: %.3f" % f_summary[0])
print(" Q1: %.3f" % f_summary[1])
print(" Median: %.3f" % f_summary[2])
print(" Q3: %.3f" % f_summary[3])
print(" Max: %.3f" % f_summary[4])
print("="*25)
```

# Seaborn 예제

### Problem 1
iris 데이터를 이용해 x축이 sepal_length이고 y축이 sepal_width인 scatter plot을 그리시오. (이때, species에 따라 점의 색이 다르게 나타나야하고, 타이틀은 'scatter plot for sepal'로 설정할 것)

### Problem 8

### Problem 10

```python
sns.catplot(penguins, x="species", y="body_mass_g", kind="box", hue="sex")
```

```python
sns.catplot(penguins, x="species", y="body_mass_g", kind="violin", hue="sex")
plt.show()
```

### Problem 11
penguins 데이터 중 species가 Adelie인 penguins에 대해서 상관계수를 구하는 코드는 다음과 같다. corr을 활용해 heatmap을 그리시오
(이때 figure의 크기는 8x8로 설정하시오)

```python
penguins_A = penguins[penguins.species=="Adelie"]

corr = penguins_A.corr()

plt.figure(figsize=(8, 8))

sns.heatmap(corr, annot=True, cmap='coolwarm',linewidth=.5)
plt.show()
```
