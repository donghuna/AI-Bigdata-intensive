```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans 
from sklearn.datasets import load_digits, load_wine
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# Digits dataset

```python
digits_dataset = load_digits()
X, y = digits_dataset.data, digits_dataset.target
print(X.shape)
```

```python
tsne = TSNE(random_state=42)
X_tsne = tsne.fit_transform(X)
```

## 시각화 (with 정답 레이블)

```python
colors = ['#476A2A', '#7851B8', '#BD3430', '#4A2D4E', '#875525',
          '#A83683', '#4E655E', '#853541', '#3A3120', '#535D8E']
plt.figure(figsize=(10, 10))
plt.xlim(X_tsne[:, 0].min(), X_tsne[:, 0].max())
plt.ylim(X_tsne[:, 0].min(), X_tsne[:, 1].max())

for i in range(len(X)):
    plt.text(X_tsne[i,0], X_tsne[i,1], str(y[i]),
    color = colors[y[i]],
    fontdict={'weight': 'bold', 'size':9})

    plt.xlabel("t-SNE feature 1")
    plt.ylabel("t-SNE feature 2")
```

## 군집화 예시

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
assignments_X = kmeans.labels_

plt.figure(figsize=(10, 10))

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=assignments_X)

plt.xlabel("t-SNE feature 1")
plt.ylabel("t-SNE feature 2")

plt.show()
```
